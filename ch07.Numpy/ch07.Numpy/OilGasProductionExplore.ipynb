```python
import pandas as pd
import numpy as np
df=pd.read_csv("OilGas1932.2014.csv")
df.head()
%matplotlib inline
```

```python
#repeat to get feel for data
df.sample(5)
```

```python
#create Series: country:value - how many times each country appears : turns out the same for all = 83
z = df['cty_name'].value_counts()

# create new DF via groupby Object on country name
df2 = df.groupby(df['cty_name']).sum()
df2 = df2.sort_values(by=['oil_prod32_14'], ascending=False)
df2['oil_prod32_14'].head(20)
#df2.head()
```

```python
# Group the dataframe by regiment, and for each regiment,
# NOTE: 'name' does not refer to the column BUT to the 'name of the group'
for name, group in df.groupby('cty_name'): 
    # print the name group
    print(name)
    # print the data of that group
    print(group['oil_prod32_14'])
```
